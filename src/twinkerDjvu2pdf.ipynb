```python
import subprocess
import pathlib

def find_similar(file):
    #print(file.name)
    parent = file.parent
    for f in parent.iterdir():
        if f.suffix != '.pdf':
            continue
        
        if file.stem in f.name:
            # print('\t', 'Exist')
            # print('\t', 'pdf: ', f.name)
            return f
    return


def gen_djvu(file):
    
    method = '.ddjvu'

    
    pdf = file.with_stem(file.stem+method).with_suffix('.pdf')
    
    inp = file.as_posix()
    out = pdf.as_posix()
    
    
    # print('\t', 'inp: ', inp)
    # print('\t', 'out: ', out)
    
    ddjvue_com = ['ddjvu', '-format=pdf', f'{inp}', f'{out}']

    process = subprocess.Popen(ddjvue_com)
    process.wait()
    
    return pdf


def gen_cmd(file):
    
    method = '.ddjvu'

    
    pdf = file.with_stem(file.stem+method).with_suffix('.pdf')
    
    inp = file.as_posix()
    out = pdf.as_posix()
    
    
    # print('\t', 'inp: ', inp)
    # print('\t', 'out: ', out)
    
    ddjvue_com = ['ddjvu', '-format=pdf', f'{inp}', f'{out}']

    return ddjvue_com
    

def get_djvu_targets(files):
    targets = []
    for file in files:
        if file.suffix not in ['.djvu', '.djv']:
            continue
        
        similar = find_similar(file)
        if similar:
            continue
        
        targets.append(file)

    return targets




```

```python
# !/usr/bin/python

import os


def getFiles(path):
    all_files = []
    for root, dirs, files in os.walk(path, topdown=False):
        for name in files:
            all_files.append(os.path.join(root, name))
    return all_files
    
def filetrByExt(files: list, exts: list = ['.djvu']):
    out = []
    
    for file in files:
        path = pathlib.Path(file)
        if path.suffix in exts:
            out.append(path.as_posix())
    
    return out
    

pa = '/Users/andrewlevin/Desktop/97üéø Tools Of Biblioteka/17üçèüèõüçéBibloteka T'
files = getFiles(pa)
files = filetrByExt(files, ['.djvu', '.djv'])
for file in files:
    makeDjvu2pdf(file)

print('\n\n'.join(files))
```
