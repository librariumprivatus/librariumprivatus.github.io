```python
import utils

HOME = '../data/biblioteka-main/'

files = utils.walk_files(HOME)
```

```python
import twinks

print('Len files: ', len(files))

djvus = list(filter(lambda file: file.suffix in ['.djv', '.djvu'], files))
print('Len djvus: ', len(djvus))

notwinks_djvus = list(filter(lambda file: not twinks.is_twink_in_same_dir(file), djvus))
print('Len notwinks_djvus: ', len(notwinks_djvus))
```
