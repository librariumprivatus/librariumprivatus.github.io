```python
import subprocess
import pathlib

from concurrent.futures import ThreadPoolExecutor
import time

def exec_(cmd):
    '''
    single command run
    command must be prepared as subprocess.call
    '''
    ret = subprocess.call(cmd)
    if ret== 0:
        print("success...")
    else:
        print("error")


def pull_run(parallel_jobs, cmds):
    '''
    run pull of jobs
    input:
        parallel_jobs: integer, how many jobs at once
        cmds: list of commands
    '''
    with ThreadPoolExecutor(max_workers=parallel_jobs) as executor:
        futures = executor.map(exec_, cmds)

def gen_comds(files):
    cmds = []
    for idx, file in enumerate(files):
        cmd = gen_cmd(file)
        cmds.append(cmd)
    return cmds


def log_task(name, data=''):
    log = root.joinpath(name)

    f = open(log.as_posix(), "w")
    f.write("Success!")
    f.write("\n")
    for row in data:
        f.write(str(row))
    f.close()


```

```python
def gen_cmd(file):
    
    
    pdf = file.with_stem(f'calibri-{file.stem}').with_suffix('.pdf')
    
    inp = file.as_posix()
    out = pdf.as_posix()
    
    #print('\t', 'inp: ', file.name)
    #print('\t', 'out: ', pdf.name)
    #print('\n\n')
    
    com = [
        'pandoc', 
        '-f', 'epub', 
        '-t', 'latex', 
        '--split-level=1', 
        '--pdf-engine=xelatex', 
        '-V', 'mainfont=\"CMU Serif\"', 
        '-o', f'{out}', 
        f'{inp}',
    ]

    
    com = [
        'ebook-convert',
        f'{inp}',
        f'{out}', 
    ]

    return com
```
